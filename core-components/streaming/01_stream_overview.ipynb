# LangChain Streaming (Python)

## Introduction

**Streaming** in LangChain allows applications to receive **real-time updates during agent execution** instead of waiting for the final response.

Streaming improves:

- User experience (UX)
- Responsiveness
- Debugging
- Real-time progress visibility

Instead of waiting for the LLM to finish generating the full response, the system **streams tokens or updates as they are generated**.

---

# What Can Be Streamed

LangChain streaming supports multiple types of runtime outputs:

| Type | Description |
|-----|-------------|
| Agent progress | Stream updates after each agent step |
| LLM tokens | Stream tokens generated by the model |
| Custom updates | Emit custom progress updates |
| Multiple modes | Combine multiple streaming modes |

These updates can be used in:

- chat interfaces
- dashboards
- logs
- monitoring systems

---

# Streaming Methods

LangChain provides two methods for streaming:

| Method | Description |
|------|-------------|
| `.stream()` | synchronous streaming |
| `.astream()` | asynchronous streaming |

Example:

```python
for chunk in agent.stream(inputs):
    print(chunk)

````python
from langchain.agents import create_agent
from langchain_anthropic import ChatAnthropic


def get_weather(city: str) -> str:
    return f"It's always sunny in {city}!"


model = ChatAnthropic(
    model="claude-haiku-4-5",
    temperature=0.1,
    max_tokens_to_sample=50,
    timeout=30,
    max_retries=3,
)


agent = create_agent(
    model=model,
    tools=[],
)

for update in agent.stream(
    {"messages": [{"role": "user", "content": "Weather in Paris?"}]},
    stream_mode="updates",
):
    print(update)
````

````python
from langchain.agents import create_agent
from langchain_anthropic import ChatAnthropic


def get_weather(city: str) -> str:
    return f"It's always sunny in {city}!"


model = ChatAnthropic(
    model="claude-haiku-4-5",
    temperature=0.1,
    max_tokens_to_sample=50,
    timeout=30,
    max_retries=3,
)

agent = create_agent(
    model=model,
    tools=[],
)

for token, metadata in agent.stream(
    {"messages": [{"role": "user", "content": "Weather in Boston"}]},
    stream_mode="messages",
):
    print(metadata["langgraph_node"])
    print(token.content_blocks)
````

````python
from langchain.agents import create_agent
from langgraph.config import get_stream_writer
from langchain_anthropic import ChatAnthropic


def get_weather(city: str) -> str:
    """
    Get the weather for a given city.
    """

    writer = get_stream_writer()

    writer(f"Searching weather data for {city}")

    return f"It's sunny in {city}"


model = ChatAnthropic(
    model="claude-haiku-4-5",
    temperature=0.1,
    max_tokens_to_sample=50,
    timeout=30,
    max_retries=3,
)


agent = create_agent(
    model=model,
    tools=[],
)

for mode, chunk in agent.stream(
    {"messages": [{"role": "user", "content": "Weather in London"}]},
    stream_mode=["updates", "custom"],
):
    print(mode, chunk)
````

````python
from langchain.agents import create_agent
from langgraph.config import get_stream_writer
from langchain_anthropic import ChatAnthropic


def search_db(query: str) -> str:
    """
    Search the database for a given query."""
    writer = get_stream_writer()

    writer("Searching database...")
    writer("Fetching records...")

    return f"Results for {query}"


model = ChatAnthropic(
    model="claude-haiku-4-5",
    temperature=0.1,
    max_tokens_to_sample=50,
    timeout=30,
    max_retries=3,
)


agent = create_agent(
    model=model,
    tools=[search_db],
)


for mode, chunk in agent.stream(
    {"messages": [{"role": "user", "content": "Find AI courses"}]},
    stream_mode=["updates", "messages", "custom"],
):
    print(mode, chunk)
````
